* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Find the minimum
tf.reduce_max(E)
```

```python
 # mean
 tf.reduce_mean(E)
```

### FIND THE MAXIMUM AND MINIMUM POSTION



```python
#Create a new tensor for finding postional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
#find the positional maxium
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F

tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Finging the min value of F

tf.reduce_min(F)
```

```python
F[tf.argmin(F)] == tf.reduce_min(F)
```
